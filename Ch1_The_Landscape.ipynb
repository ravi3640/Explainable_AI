# Section 1.1 Examples of What Explainable AI Is
Explainable AI (aka XAI) is a difficult term to define as it is difficult to include all the required angles in the definition. But basic defintion is XAI is a set of methods and tools that can be adopted to make ML models understandable to human. 

**Three fundamental aspects of XAI**


1.   **Learning Phase:** Train Convolutional Neural Network(CNN) model to classify images into dogs and wolf. Model achieved a remarkable accuracy as there is only one miscalssification (model miscassified a dog image as wolf). Lets use an XAI method and ask the model, "why have you classified it as wolf?". It answers as there is snow in the picture and wolfs are more likely to be found in snow than dogs. ML expert will correct the error by training the model on data where part if image is covered.

    **Learning-** Accuracy is not enough to say that a model is a good model. If model povides bad explanation, them model can fail us systematically.  
2.   **Knowledge Discovery:** XAI can help in understanding the model's decision making process. Which can help in answering questions like - "Is the model fair?" 

   **Learning-** Check for bias as this is one of the main focus points of XAI.
3. **Reliability and Robustness:** Deconstruct the logic of the model to identify what are the factors that plays key role in model decision making. If model can be fooled by copying those key attributes than XAI cab be used to detect such problems. 

   **Learning-** Check model and make it robust against malicious attack

**In light of above three fundamentals, XAI tries to make ML system more transparent and interpretable to build trust and confidence in its adoption.**
